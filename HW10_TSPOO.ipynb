```python
# Let's Make A Deal - Monty Hall
# 3 Doors

#initial paramters
num_noswitch = 0 #number of times the player did not switch
num_switch = 0   #number of times the player did switch
NSW = 0 # no switch win
NSL = 0 # no switch lost
SW = 0  # switch win
SL = 0  # switch lost
nvals = np.arange(0,1000,1)

for n in nvals:
    prize = np.random.randint(1,4) #end term is exclusive, so this would give me three doors
    player = np.random.randint(1,4)
    #print(prize,player)
    dooropen = np.random.randint(1,4)
    if dooropen==prize or dooropen==player:
        while dooropen==prize or dooropen==player: dooropen = np.random.randint(1,4)
        decide = np.random.random()
        if decide < 0.5:
            #switch
            num_switch+=1
            if player==prize:SL+=1
            else: SW+=1
        else:
            #no switch
            num_noswitch+=1
            if player==prize: NSW+=1
            else: NSL+=1    
    else:
        decide = np.random.random()
        if decide < 0.5:
            #switch
            num_switch+=1
            if player==prize:SL+=1
            else: SW+=1
        else:
            #no switch
            num_noswitch+=1
            if player==prize: NSW+=1
            else: NSL+=1
                
print("Number of NO switch=", num_noswitch)
print("Number of switch=" ,num_switch)
print("Number of NO switch Wins=" ,NSW)
print("Number of NO switch Loses=" ,NSL)
print("Number of switch wins=",SW)
print("Number of switch loses=",SL)

print("Probalility of switch winning=",SW/(num_switch))
print("Prob of no switch winning=", NSW/(num_noswitch))
```

```python
# 4 Doors

#initial paramters
num_noswitch = 0 #number of times the player did not switch
num_switch = 0   #number of times the player did switch
NSW = 0 # no switch win
NSL = 0 # no switch lost
SW = 0  # switch win
SL = 0  # switch lost
nvals = np.arange(0,1000,1)

for n in nvals:
    prize = np.random.randint(1,5) #end term is exclusive, so this would give me four doors
    player = np.random.randint(1,5)
    #print(prize,player)
    dooropen = np.random.randint(1,5)
    if dooropen==prize or dooropen==player:
        while dooropen==prize or dooropen==player:dooropen = np.random.randint(1,5)
        decide = np.random.random()
        if decide < 0.5:
            #switch
            num_switch+=1
            switch = np.random.random()
            if player==prize: SL+=1
            elif switch < 0.5: SL+=1
            else: SW+=1
        else:
            #no switch
            num_noswitch+=1
            if player==prize: NSW+=1
            else: NSL+=1   
    else:
        decide = np.random.random()
        if decide < 0.5:
            #switch
            num_switch+=1
            switch = np.random.random()
            if player==prize: SL+=1
            elif switch < 0.5: SL+=1
            else: SW+=1
        else:
            #no switch
            num_noswitch+=1
            if player==prize: NSW+=1
            else: NSL+=1
                
print("Number of NO switch=", num_noswitch)
print("Number of switch=" ,num_switch)
print("Number of NO switch Wins=" ,NSW)
print("Number of NO switch Loses=" ,NSL)
print("Number of switch wins=",SW)
print("Number of switch loses=",SL)

print("Probalility of switch winning=",SW/(num_switch))
print("Prob of no switch winning=", NSW/(num_noswitch))
```
